```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
claim_dataset['authorities_contacted']=label_encoder.fit_transform(claim_dataset["authorities_contacted"])
```

```python
claim_dataset=claim_dataset.drop('_c39',axis=1)
```

```python
claim_dataset=claim_dataset.astype({"policy_annual_premium":'str'})

print(claim_dataset.dtypes)
```

```python
claim_dataset.select_dtypes(include="object").columns
```

```python
claim_dataset['policy_state']=label_encoder.fit_transform(claim_dataset["policy_state"])
claim_dataset['policy_csl']=label_encoder.fit_transform(claim_dataset["policy_csl"])
claim_dataset['insured_sex']=label_encoder.fit_transform(claim_dataset["insured_sex"])
claim_dataset['insured_education_level']=label_encoder.fit_transform(claim_dataset["insured_education_level"])
claim_dataset['insured_occupation']=label_encoder.fit_transform(claim_dataset["insured_occupation"])
claim_dataset['insured_hobbies']=label_encoder.fit_transform(claim_dataset["insured_hobbies"])
claim_dataset['insured_relationship']=label_encoder.fit_transform(claim_dataset["insured_relationship"])
claim_dataset['incident_type']=label_encoder.fit_transform(claim_dataset["incident_type"])
claim_dataset['collision_type']=label_encoder.fit_transform(claim_dataset["collision_type"])
claim_dataset['incident_severity']=label_encoder.fit_transform(claim_dataset["incident_severity"])
claim_dataset['incident_state']=label_encoder.fit_transform(claim_dataset["incident_state"])
claim_dataset['incident_city']=label_encoder.fit_transform(claim_dataset["incident_city"])
claim_dataset['incident_location']=label_encoder.fit_transform(claim_dataset["incident_location"])
claim_dataset['property_damage']=label_encoder.fit_transform(claim_dataset["property_damage"])
claim_dataset['police_report_available']=label_encoder.fit_transform(claim_dataset["police_report_available"])
claim_dataset['auto_make']=label_encoder.fit_transform(claim_dataset["auto_make"])
claim_dataset['auto_model']=label_encoder.fit_transform(claim_dataset["auto_model"])
claim_dataset['fraud_reported']=label_encoder.fit_transform(claim_dataset["fraud_reported"])
```

```python
claim_dataset.info()
```

```python
claim_dataset.shape
```

```python
claim_dataset.head()
```

```python
X=claim_dataset.drop(['fraud_reported','policy_bind_date','incident_date'],axis=1)
Y=claim_dataset['fraud_reported']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
X_test_prediction = model.predict(X_test)

data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy:', data_accuracy)
```
